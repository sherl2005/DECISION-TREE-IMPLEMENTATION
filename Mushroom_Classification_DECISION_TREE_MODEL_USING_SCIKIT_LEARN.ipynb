```python
# Load libraries
import pandas as pd

from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
df = pd.read_csv("/content/mushrooms.csv")# load dataset
df.head()
```

```python
df.isnull().sum().sum()
#check if there are any missing values
```

```python
df.duplicated().sum()
# number of duplicate rows
```

```python
# Define target and feature variables
target_variable = "class"  # Dependent variable
feature_variables = [col for col in df.columns if col != target_variable]  # Independent variables

# Creating separate DataFrames for more clarity
X = df[feature_variables]  # Features
y = df[target_variable]  # Target
```

```python
# Encoding categorical features and target variable
en_X = X.apply(lambda col: col.astype("category").cat.codes)  # categorical to numerical
en_y = y.astype("category").cat.codes  #target to numerical
```

```python
# Split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(en_X, en_y, test_size=0.2, random_state=42)#80% training, 20% testing
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
